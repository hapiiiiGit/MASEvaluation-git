```
import warnings
import logging

root_logger = logging.getLogger()
root_logger.setLevel(logging.ERROR)
warnings.filterwarnings('ignore')
```

```
def max_average_path_cost(matrix):
    n = len(matrix)
    # dp will store tuples of (max_total_cost, path_length) for each cell
    dp = [[(float('-inf'), 0) for _ in range(n)] for _ in range(n)]
    dp[0][0] = (matrix[0][0], 1)
    
    for i in range(n):
        for j in range(n):
            if i == 0 and j == 0:
                continue
            candidates = []
            if i > 0:
                prev_cost, prev_len = dp[i-1][j]
                candidates.append((prev_cost + matrix[i][j], prev_len + 1))
            if j > 0:
                prev_cost, prev_len = dp[i][j-1]
                candidates.append((prev_cost + matrix[i][j], prev_len + 1))
            # Choose the candidate with the maximum average cost
            best = max(candidates, key=lambda x: x[0]/x[1])
            dp[i][j] = best
    
    max_total_cost, path_length = dp[-1][-1]
    return max_total_cost / path_length
```
