```
import warnings
import logging

root_logger = logging.getLogger()
root_logger.setLevel(logging.ERROR)
warnings.filterwarnings('ignore')
```

```
import pandas as pd

def load_file(file_path):
    """
    Load a CSV or Excel file into a pandas DataFrame.
    Handles errors gracefully and returns None if loading fails.
    """
    try:
        if file_path.lower().endswith('.csv'):
            df = pd.read_csv(file_path)
        elif file_path.lower().endswith(('.xls', '.xlsx')):
            df = pd.read_excel(file_path)
        else:
            print(f"Unsupported file format: {file_path}")
            return None
        print(f"Loaded file: {file_path} (shape: {df.shape})")
        return df
    except Exception as e:
        print(f"Error loading file {file_path}: {e}")
        return None
```

```
def clean_and_save_data(df, output_path='cleaned_output.csv'):
    """
    Clean the DataFrame by removing duplicates, handling missing values,
    standardizing column names, and saving to a new CSV file.
    """
    if df is None:
        print("No data to clean.")
        return

    # Standardize column names: lowercase and replace spaces with underscores
    df.columns = [col.strip().lower().replace(' ', '_') for col in df.columns]

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Handle missing values: fill numeric columns with 0, others with empty string
    for col in df.columns:
        if df[col].dtype.kind in 'biufc':  # numeric types
            df[col] = df[col].fillna(0)
        else:
            df[col] = df[col].fillna('')

    # Save cleaned data to CSV
    try:
        df.to_csv(output_path, index=False)
        print(f"Cleaned data saved to {output_path} (shape: {df.shape})")
    except Exception as e:
        print(f"Error saving cleaned data: {e}")
```
