```
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_data(urls, data_types):
    """
    Scrape specified data types from a list of URLs.
    Args:
        urls (list): List of URLs to scrape.
        data_types (list): List of data types to extract, e.g., ['product', 'price', 'review'].
    Returns:
        pd.DataFrame: DataFrame containing scraped data.
    """
    results = []
    for url in urls:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        data = {'url': url}
        # Example extraction logic (should be customized per site/data type)
        if 'product' in data_types:
            product = soup.find('h1') or soup.find('title')
            data['product'] = product.text.strip() if product else None
        if 'price' in data_types:
            price = soup.find(class_='price') or soup.find(string=lambda t: '$' in t)
            data['price'] = price.text.strip() if hasattr(price, 'text') else (price.strip() if price else None)
        if 'review' in data_types:
            review = soup.find(class_='review') or soup.find('p')
            data['review'] = review.text.strip() if review else None
        results.append(data)
    return pd.DataFrame(results)

# Example usage:
urls = [
    # Add URLs to scrape here
]
data_types = ['product', 'price', 'review']
df = scrape_data(urls, data_types)
df.to_csv('scraped_sales_data.csv', index=False)
```
