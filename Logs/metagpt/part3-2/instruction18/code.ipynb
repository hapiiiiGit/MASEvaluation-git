```
import torch
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import numpy as np

# Assume 'model' is the trained PyTorch model from previous steps
# Assume 'test_loader' is the DataLoader for the test set, processed identically to training data

model.eval()
all_labels = []
all_preds = []
all_scores = []

with torch.no_grad():
    for data, labels in test_loader:
        outputs = model(data)
        # If outputs are logits, apply sigmoid for binary classification
        scores = torch.sigmoid(outputs).cpu().numpy().flatten()
        preds = (scores > 0.5).astype(int)
        all_labels.extend(labels.cpu().numpy().flatten())
        all_preds.extend(preds)
        all_scores.extend(scores)

# Calculate metrics
accuracy = accuracy_score(all_labels, all_preds)
precision = precision_score(all_labels, all_preds)
recall = recall_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds)
roc_auc = roc_auc_score(all_labels, all_scores)

# Generate report
report = f"""
Model Performance Report for Anomaly Detection (Loitering & Erratic Movement):

Accuracy:  {accuracy:.4f}
Precision: {precision:.4f}
Recall:    {recall:.4f}
F1-score:  {f1:.4f}
ROC-AUC:   {roc_auc:.4f}

The model was evaluated on the test set using standard metrics for anomaly detection. 
Higher precision indicates fewer false alarms, while higher recall means more anomalies are detected.
F1-score balances both, and ROC-AUC reflects overall discrimination ability.

"""

print(report)
```
