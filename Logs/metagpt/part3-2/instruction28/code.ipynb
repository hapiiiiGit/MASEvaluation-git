```
from metagpt.tools.libs.terminal import Terminal
terminal = Terminal()
await terminal.run_command('pip install pandas requests')

import pandas as pd
import requests
from io import StringIO

# Step 1: Download the WHO COVID-19 global data CSV using requests with headers
who_covid_url = "https://covid19.who.int/WHO-COVID-19-global-data.csv"
headers = {
    "User-Agent": "Mozilla/5.0"
}
response = requests.get(who_covid_url, headers=headers)
response.raise_for_status()  # Raise an error if the request failed

# Step 2: Load the CSV content into a pandas DataFrame, skipping bad lines
csv_data = StringIO(response.text)
df = pd.read_csv(csv_data, delimiter=',', on_bad_lines='skip')

# Step 3: Display the first few rows to confirm successful extraction
print("First few rows of the dataset:")
print(df.head())

# Step 4: Basic cleaning - check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Step 5: Show basic info for further analysis
print("\nDataFrame info:")
print(df.info())

# Step 6: Show column names for further analysis
print("\nColumn names:")
print(df.columns.tolist())
```

```
# The previously loaded data is invalid (HTML content, not a healthcare dataset).
# Let's download a valid healthcare dataset (e.g., diabetes) from the UCI Machine Learning Repository.

import pandas as pd

# UCI Diabetes dataset (Pima Indians Diabetes Database)
uci_diabetes_url = "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"

# Load the dataset into a pandas DataFrame
df_raw = pd.read_csv(uci_diabetes_url)
df = df_raw.copy()

# Display basic info for cleaning
print("First few rows of the dataset:")
print(df.head())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nData types per column:")
print(df.dtypes)
print("\nColumn names:")
print(df.columns.tolist())
```

```
from metagpt.tools.libs.data_preprocess import get_column_info

column_info = get_column_info(df)
print("column_info")
print(column_info)
```

```
import numpy as np

# Columns to clean (cannot be zero physiologically)
cols_to_clean = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Copy the DataFrame before processing
df_clean = df.copy()

# Replace zero values with NaN in specified columns
df_clean[cols_to_clean] = df_clean[cols_to_clean].replace(0, np.nan)

# Impute missing values with the median of each column
for col in cols_to_clean:
    median_val = df_clean[col].median()
    df_clean[col].fillna(median_val, inplace=True)

# Show summary after cleaning
print("Summary after cleaning:")
print(df_clean[cols_to_clean].describe())
print("\nMissing values per column after cleaning:")
print(df_clean.isnull().sum())
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical summary for numeric columns
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
print("Statistical summary for numeric columns:")
print(df_clean[numeric_cols].describe())

# Correlation matrix
corr_matrix = df_clean[numeric_cols].corr()
print("\nCorrelation matrix:")
print(corr_matrix)

# Visualizations
plt.figure(figsize=(16, 10))

# Histograms for key features
for i, col in enumerate(['Glucose', 'BloodPressure', 'BMI', 'Age']):
    plt.subplot(2, 2, i+1)
    sns.histplot(df_clean[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()

# Boxplots for key features
plt.figure(figsize=(16, 10))
for i, col in enumerate(['Glucose', 'BloodPressure', 'BMI', 'Age']):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x=df_clean[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# Outcome distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df_clean)
plt.title('Distribution of Diabetes Outcome')
plt.xlabel('Outcome (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Count')
plt.show()
```
