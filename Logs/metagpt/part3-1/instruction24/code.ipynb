```
from googleapiclient.discovery import build

def setup_youtube_client(api_key: str):
    """
    Set up the YouTube Data API client using the provided API key.
    
    Args:
        api_key (str): Your YouTube Data API key.
    
    Returns:
        Resource: A resource object for interacting with the YouTube Data API.
    """
    youtube = build('youtube', 'v3', developerKey=api_key)
    return youtube

# Example usage
# api_key = 'YOUR_API_KEY'  # Replace with your actual API key
# youtube_client = setup_youtube_client(api_key)
```

```
def search_videos(youtube, query: str, max_results: int = 50):
    """
    Search for videos on YouTube based on a query.
    
    Args:
        youtube (Resource): The YouTube Data API client.
        query (str): The search term or channel name.
        max_results (int): The maximum number of results to return.
    
    Returns:
        list: A list of video details.
    """
    request = youtube.search().list(
        q=query,
        part='id,snippet',
        maxResults=max_results,
        publishedAfter='2023-10-01T00:00:00Z'  # Adjust this date to get videos from the last 10 days
    )
    response = request.execute()
    
    videos = []
    for item in response.get('items', []):
        if item['id']['kind'] == 'youtube#video':
            videos.append({
                'title': item['snippet']['title'],
                'videoId': item['id']['videoId'],
                'publishedAt': item['snippet']['publishedAt'],
                'description': item['snippet']['description'],
                'channelTitle': item['snippet']['channelTitle']
            })
    
    return videos
```
