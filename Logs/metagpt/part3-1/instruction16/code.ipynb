```
import pandas as pd
import numpy as np

# Define the number of samples
num_samples = 1000

# Define fruit types and their corresponding freshness labels
fruit_types = ['Apple', 'Banana', 'Orange', 'Grapes', 'Strawberry']
freshness_labels = ['Fresh', 'Ripe', 'Overripe']

# Generate synthetic data
data = {
    'Fruit': np.random.choice(fruit_types, num_samples),
    'Freshness': np.random.choice(freshness_labels, num_samples),
    'Days_to_Expiry': np.random.randint(0, 10, num_samples)  # Days remaining before expiry
}

# Create a DataFrame
mock_dataset = pd.DataFrame(data)

# Display the first few rows of the mock dataset
mock_dataset.head()
```
