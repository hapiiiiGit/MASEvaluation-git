```
from metagpt.tools.libs.terminal import Terminal
terminal = Terminal()

# Create a new Scrapy project named 'data_scraper'
project_name = "data_scraper"
spider_name = "sample_spider"
sample_website = "http://example.com"

# Terminal commands to create the Scrapy project and spider
commands = f"""
scrapy startproject {project_name}
cd {project_name}
scrapy genspider {spider_name} {sample_website}
"""

# Execute the commands in the terminal
await terminal.run_command(commands)
```

```
from metagpt.tools.libs.web_scraping import view_page_element_to_scrape

# View the HTML content of the sample website to understand its structure
html_content = await view_page_element_to_scrape(sample_website, "Extract specific data from the website", keep_links=False)

# Further code for parsing HTML content using BeautifulSoup and extracting specific data will be added here
```

```
from bs4 import BeautifulSoup
import pandas as pd

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Example: Extracting all paragraph texts from the sample website
data = [p.get_text() for p in soup.find_all('p')]

# Create a DataFrame from the extracted data
df = pd.DataFrame(data, columns=['Paragraph Text'])

# Display the DataFrame
df.head()
```
